```python
class File:
    def __init__(self):
        self._ma_file = []
        
    def __len__(self):
        return len(self._ma_file)
    
    def __str__(self):
        return "<" + str(self._ma_file)[1:-1] + "<"
    
    def premier(self):
        if len(self) > 0:
            return self._ma_file[0]
        
    def enfiler(self, e):
        self._ma_file.append(e)
        
    def defiler(self):
        if len(self) > 0:
            return self._ma_file.pop(0)
        
        
        

f = File()
f.enfiler(3)
f.enfiler(5)
print(f.premier())
f.enfiler(8)
print(f)
print(len(f))
f.defiler()
print(f.defiler())
print(f.defiler())
```
